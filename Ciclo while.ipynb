### Peso de ratones, acumulado y cantidad debajo de un minimo

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main() {
    int p, pa = 0, c = 0;
    srand(time(NULL));
    
    while (pa <= 4000){
        p = (rand() % 800) + 100;
        printf("El raton pesa %d gramos\n", p);
        pa += p;
        if (p < 600){
            c++;
        }
    }
    printf("El peso acumulado de los ratones es %d gramos\n", pa);
    printf("Hay %d ratones que pesan menos de 600 gramos", c);
}
```

### Peso de ratones V2

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main(){
    int p, pa = 0, c = 0, rl = 99999, rp = -99999;
    srand(time(NULL));
    
    while (pa <= 4000){
        p = (rand() % 800) + 100;
        printf("El raton pesa %d gramos\n", p);
        pa += p;
        if (p < 600){
            c++;
        }
        if (p < rl){
            rl = p;
        }
        if (p > rp){
            rp = p;
        }
    }
    printf("Hay %d ratones que pesan menos de 600 gramos\n", c);
    printf("El peso acumulado de los ratones es %d gramos\n", pa);
    printf("El raton mas liviano pesa %d gramos\n", rl);
    printf("El raton mas pesado pesa %d gramos\n", rp);
}
```

### Promedio de edad con menores

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main(){
    float pr = 0;
    int m = 0, t = 0, e;
    srand(time(NULL));
    
    while (m < 5){
        e = (rand() % 50) + 1;
        printf("Edad \t %d\n", e);
        t++;
        pr += e;
        if (e < 18){
            m++;
        }
    }
    printf("El promedio de edad es %f\n", pr/t);
    printf("Hay %d mayores de edad", t-m);
}
```

### Calculos con personas

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main(){
    int m = 0, e, me = 0, mme = 0;
    int h = 0, sh = 0;
    char g;
    srand(time(NULL));
    
    while (m < 5){
        e = (rand() % 50) + 1;
        if ((rand() % 2) == 0){
            g = 'M';
        }
        else {
            g = 'F';
        }
        printf("Edad:\t %d\t Genero:\t %c\n", e, g);
        if (e >= 18){
            me++;
            if (g == 'F'){
                mme++;
                m++;
            }
            else{
                h++;
                sh += e;
            }
        }else{
            if (g == 'F'){
                m++;
            }
            else{
                h++;
                sh += e;
            }
        }
    }
    printf("Hay %d mayores de edad\n", me);
    printf("Hay %d mujeres mayores de edad\n", mme);
    printf("Hay un total de %d hombres\n", h);
    printf("Los hombres suman una edad de %d", sh);
}
```

### Vacas
Las vacas pesan en promedio 540Kg, tomaremos una distribucion de 300-780 Kg

```c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int main(){
    int v, nv = 0, pv = 0;
    srand(time(NULL));
    
    while (pv < 40000){
        v = (rand() % 480) + 300;
        if ((v + pv) < 40000){
            printf("Agregando vaca de peso %d\n", v);
            pv += v;
            nv++;
        }
        else {
            break;
        }
    }
    printf("Se cargaron en el camion %d vacas por un peso de %d kilogramos", nv, pv);
}
```

El rango de temperatura sera -20 y 65
